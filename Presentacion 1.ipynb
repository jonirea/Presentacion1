```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("./data/top100.csv")
```

```python
df.sort_values(by=['popularity'], ascending=False)
```

```python
top_artists = df["artist"].value_counts().head(10).to_frame()
top_artists
```

```python
top_artists['Index'] = range(0, 0+len(top_artists))

top_artists = top_artists.reset_index().set_index('Index')

top_artists = top_artists.rename(columns = {"index" : "Artist", "artist" : "Contador"})
```

```python
artists = top_artists["Artist"]

app = top_artists["Contador"]

plt.rcParams["figure.figsize"] = (15,5)

plt.subplot().set_title('Artistas Top', fontsize = 25)

plt.xticks(fontsize = 10, rotation=-20)

plt.yticks(fontsize =8)

plt.bar(artists, app)

plt.xlabel("Artistas", fontsize = 15)

plt.ylabel("Numero de canciones en el top 100", fontsize = 15)

plt.show()
```

```python
top_genres = df["top genre"].value_counts().head(10).to_frame()
top_genres
```

```python
top_genres['Index'] = range(0, 0+len(top_artists))

top_genres = top_genres.reset_index().set_index('Index')

top_genres = top_genres.rename(columns = {"index" : "Genre", "top genre" : "Contador"})
```

```python
genres = top_genres["Genre"]

app = top_genres["Contador"]

plt.rcParams["figure.figsize"] = (15,5)

plt.subplot().set_title("Generos Top", fontsize = 20)

plt.bar(genres, app)

plt.yticks(fontsize = 8)

plt.xticks(fontsize = 10, rotation=-75)

plt.xlabel("Genero", fontsize = 15)

plt.ylabel("Numero de canciones en el top 100", fontsize = 15)

plt.show()
```
